```python
# Notebook: 8_Cluster_Profiling.ipynb

import pandas as pd

# 1. Load the final clustered data
df = pd.read_csv('final_clustered_netflix_data.csv')

# 2. Analyze the top genres for each cluster
print("--- Cluster Profiles (Top Genres) ---")

for i in range(6): # Assuming 6 clusters
    print(f"\nCluster {i} Profile:")
    
    # Getting all rows in this cluster
    cluster_data = df[df['cluster'] == i]
    
    # Show the top 3 most frequent genres
    top_genres = cluster_data['listed_in'].value_counts().head(3)
    print("Top Genres:")
    print(top_genres)
    
    # Show 5 sample titles to see if the model working fine
    print("Sample Titles:")
    print(cluster_data['title'].head(5).tolist())
    print("-" * 30)

# 3. Saving summary for report
cluster_summary = df.groupby('cluster')['listed_in'].agg(lambda x: x.value_counts().index[0]).reset_index()
cluster_summary.columns = ['Cluster', 'Dominant Genre']
print("\n--- Summary Table ---")
print(cluster_summary)
```
